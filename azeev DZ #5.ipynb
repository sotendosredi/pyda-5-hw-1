```python
documents = [ 
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'}, 
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'}, 
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'} 
]

print(type(documents[0]['number']))

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

functions = {'p':'p_owner_name',
             'l':'list_doc',
             's':'num_shell',
             'as':'add_shell',
             'd':'del_doc',
             'm':'move_doc',
             'a':'add_doc'}

input_num = '11-2'



# функция работы с библиотекой
def start_lib():
    
    input_comand = input('Введите команду: ')
 
    if input_comand == 'p':
        p_owner_name()
    elif input_comand == 'l':
        list_doc()
    elif input_comand == 's':
        num_shell()
    elif input_comand == 'as':
        add_shell()
    elif input_comand == 'd':
        del_doc()
    elif input_comand == 'm':
        move_doc()
    elif input_comand == 'a':
        add_doc()
    else:
        print('Вы ввели неверную команду, попробуйте снова')

# функция, которая выдает имя владельца номера документа
def p_owner_name():
    input_num = input('Введите номер документа: ')
    doc_elem = []

    for elem in documents:
        doc_elem += elem.values()
        
    if  input_num in doc_elem:
        print(doc_elem[(doc_elem.index(input_num)) + 1])
    else:
        print('Владелец с указанным номером документа в базе отсутствует или вы указали несуществующий номер документа')

# функция вывода списка    
def list_doc():
    for i in range(len(documents)):
        pr = ''
        for key in documents[i]:    
            if key == list(documents[i].keys())[0]:
                pr = pr + list(documents[i].values())[0]
            else:
                for a in range(len((documents[i].values()))):
                    if a == 0:
                        continue
                    else:
                        pr = pr + ' ' + '"' + list(documents[i].values())[a] + '"'
                break
        print('\033[1m' + pr)

# функция возврата полки
def num_shell():
    input_num = input('Введите номер документа: ')   
    for i in range(len(directories)):
        if input_num in list(directories.values())[i]:
            return list(directories.keys())[i]
        else:
            print('Вы указали несуществующий или неверный номер документа')
            break

# функция добавления полки 
def add_shell(): 
    input_new_shell = input('Введите номер полки, которую вы желаете добавить: ')
    if input_new_shell in directories.keys():
        print('Полка с таким номером уже существует')
    else:
        directories.update({input_new_shell:[]})

        


# функция для удаления номера документа с directories and documents    
def del_doc():
    input_num = str(input('Введите номер документа: '))

    # удаление из directories
    a_elem = []
    for elem in directories.values():
        a_elem += elem
    if input_num in a_elem:
        for key, val in directories.items():
            if input_num in val:
                directories[key].remove(input_num)           
    else:
        print('Вы указали несуществующий или неверный номер документа')

# удаление из documents
    for i in range(len(documents)):
        if input_num == documents[i]['number']:
            documents.pop(i)
            break
        else:
            continue     


# перемещение документа по directories
def move_doc():
    input_num = input('Введите номер документа: ')
    input_num_shell = input('Введите номер полки, на которую следует переместить документ: ')

    for i in range(len(list(directories.keys()))):
        if str(input_num) in directories[list(directories.keys())[i]]: 
            directories[list(directories.keys())[i]].remove(str(input_num))
        else:
            print('Вы указали несуществующий или неверный номер документа')
            break
        if input_num_shell in list(directories.keys()):
            directories[str(input_num_shell)].append(input_num)
            break
        else:
            print('Вы указали несуществующий номер полки')
            break


# функция добавления элемента в библиотеку
def add_doc():
    input_type = input('Введите вид документа: ')
    input_number = input('Введите номер документа: ')
    input_name = input('Введите име владельца документа: ')
    input_num_shell = input('Введите номер полки, на которую следует переместить документ: ')

    doc_dict = {'type':None, 'number':None, 'name':None}

    if input_num_shell not in directories.keys():
        print('Указанная полка отсутвует, выберите другую полку или создайте новую, используя команду -as')
    else:
        doc_dict['type'] = input_type
        doc_dict['number'] = input_number
        doc_dict['name'] = input_name
        documents.append(doc_dict)

        # добавление документа на полку
        for i in range(len(list(directories.keys()))):
            if input_num_shell in list(directories.keys()):
                    directories[str(input_num_shell)].append(input_number)
                    break
            else:
                print('Вы указали несуществующий номер полки')
                break

start_lib()
```
